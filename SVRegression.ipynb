```python
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
import tensorflow
tensorflow.keras.__version__
```

```python
weather = pd.read_csv('sonomavalley.csv')
weather.head()
```

```python
weatherClean=weather.drop("Unnamed: 0", axis=1)
weatherClean.head()
```

```python
# Data Pre-Processing
```

```python
weatherClean.columns
```

```python
weatherCleandf = weatherClean.drop(['winery', 'vintage', 'cities'], axis=1)
```

```python
X = weatherCleandf.drop(['points'], axis=1).values.astype(float)
y = weatherCleandf['points'].values.astype(float).reshape(-1, 1)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
regressor = SVR(kernel='rbf')
regressor.fit(X,y)
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Support Vector Regression Model(High Resolution))')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
