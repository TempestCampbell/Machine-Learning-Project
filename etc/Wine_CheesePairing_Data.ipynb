```python
import pandas as pd
from pandas import DataFrame
import lxml 
from pandas import Series, DataFrame

```

```python
df2=pd.read_html("https://www.gourmetfoodstore.com/wine-cheese-15193", flavor= "html5lib")
# df2
```

```python
# df2[0]
```

```python
#Cheese Pairing by Type of Wine
df2_Wine=pd.DataFrame(df2[0])
df2_Wine
```

```python
#Rename columns
df2_Wine.columns={"Type of Wine":"0","Wine Examples":"1","Cheese Type":"2", "Cheese Examples":"3"}
df2_Wine
```

```python
#Delete row 0 
df2_Wined=df2_Wine.drop(df2_Wine.index[0])
df2_Wined
```

```python
#Turn the DF into a CSV
df2_Wined.to_csv("TypeofWinetoCheese")
```

```python
#Split by Type of Wine
s = df2_Wined['Type of Wine'].str.split(',').apply(Series, 1).stack()
s.index = s.index.droplevel(-1) # to line up with df's index
s.name = 'Type of Wine' # needs a name to join
del df2_Wined['Type of Wine']
SplitWine=df2_Wined.join(s)
SplitWine
```

```python
SplitWine
```

```python
c = SplitWine['Cheese Examples'].str.split(',').apply(Series, 1).stack()
c.index = c.index.droplevel(-1) # to line up with df's index
c.name = 'Cheese Examples' # needs a name to join
del SplitWine['Cheese Examples']
SplitCheeseEx= SplitWine.join(c)
SplitCheeseEx
```

```python
c = SplitCheeseEx['Cheese Type'].str.split(',').apply(Series, 1).stack()
c.index = c.index.droplevel(-1) # to line up with df's index
c.name = 'Cheese Type' # needs a name to join
del SplitCheeseEx['Cheese Type']
SplitCheeseType= SplitCheeseEx.join(c)
SplitCheeseType
```

```python
c = SplitCheeseType['Wine Examples'].str.split(',').apply(Series, 1).stack()
c.index = c.index.droplevel(-1) # to line up with df's index
c.name = 'Wine Examples' # needs a name to join
del SplitCheeseType['Wine Examples']
SplitWineEx= SplitCheeseType.join(c)
SplitWineEx
```
